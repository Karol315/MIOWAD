```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from nn.network import SimpleNetwork
from nn.metrics import mse
```

```python
train_data = pd.read_csv('../datasets/regression/square-simple-training.csv')


x_train = train_data['x']
y_train = train_data['y']


x_train_min = x_train.min()
x_train_max = x_train.max()
y_train_min = y_train.min()
y_train_max = y_train.max()

x_train_normalized = (x_train - x_train_min) / (x_train_max - x_train_min)
```

```python
network = SimpleNetwork(
    input_size=1,
    hidden_layers_sizes=[5, 5],
    output_size=1
)
```

```python
y_train_pred = []
for i in range(len(train_data['x'])):
    y_train_pred.append(
        network.predict(inputs = x_train_normalized[i])[0] * (y_train_max - y_train_min) + y_train_min
                  )
plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, c ='blue')
plt.scatter(x_train, y_train_pred, c ='red')
plt.show()
```
