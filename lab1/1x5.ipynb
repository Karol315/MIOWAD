```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from nn.network import SimpleNetwork
from nn.metrics import mse
```

```python
train_data = pd.read_csv('../datasets/regression/square-simple-training.csv')

#normalizacja
x = train_data['x']
y = train_data['y']


x_min = x.min()
x_max = x.max()
y_min = y.min()
y_max = y.max()

x_normalized = (x - x_min) / (x_max - x_min)
y_normalized = (y - y_min) / (y_max - y_min)
```

```python
network = SimpleNetwork(
    input_size=1,
    hidden_layers_sizes=[5],
    output_size=1
)
```

```python
y_pred = []
for i in range(len(train_data['x'])):
    y_pred.append(
        network.predict(inputs = x_normalized[i])[0] * (y_max - y_min) + y_min
                  )
plt.figure(figsize=(10, 6))
plt.scatter(train_data['x'], train_data['y'], c ='blue')
plt.scatter(train_data['x'], y_pred, c ='red')
plt.show()
```

```python
network.hidden_layers[0].set_weights([   [-17.041677 ],
                                         [-42.277992 ],
                                         [-12.716685 ],
                                         [ -2.9805634],
                                         [ 35.923325 ]])

network.hidden_layers[0].set_bias([ 11.1751,  43.3852,  11.2032,   1.6433, -19.0130])

network.output_layer.set_weights([[ -4.628575,  -10.532224,   -6.708138,   20.003542,    3.5743825]])

network.output_layer.set_bias( [5.4565])
```

```python
y_pred = []
for i in range(len(train_data['x'])):
    y_pred.append(
        network.predict(inputs = x_normalized[i])[0] * (y_max - y_min) + y_min
                  )
plt.figure(figsize=(10, 6))
plt.scatter(train_data['x'], train_data['y'], c ='blue')
plt.scatter(train_data['x'], y_pred, c ='red')
plt.show()
```

```python
mse(y, y_pred)
```
