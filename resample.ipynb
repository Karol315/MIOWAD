```python
## ze względu na niereprezentatywny podział w sqare large między train i test zdecydowałem się zrobić resampling
```

```python
import pandas as pd

df_train = pd.read_csv('datasets/regression/square-large-training.csv')
df_test = pd.read_csv('datasets/regression/square-large-test.csv')
```

```python
df_test.shape
```

```python
df_train.shape
```

```python
df_combined = pd.concat([df_train, df_test], ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split

df_train_new, df_test_new = train_test_split(
    df_combined,
    train_size=1000,
    test_size=300,
    random_state=42
)
```

```python
df_test_new = pd.DataFrame(df_test_new)
df_train_new = pd.DataFrame(df_train_new)
```

```python
df_test_new.to_csv('datasets/regression/my-square-large-test.csv', index=False)
df_train_new.to_csv('datasets/regression/my-square-large-training.csv', index=False)
```
